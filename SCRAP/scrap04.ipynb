```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://alexa.com/topsites/countries/KR'
res = requests.get(url)
res
```

```python
html_str = res.text
soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
div_list = soup.find_all(class_='site-listing')
len(div_list)
```

```python
div_list[0]
```

```python
div_list[0].select_one('div.DescriptionCell > p > a').text
```

```python
div_list[0].select_one('div.DescriptionCell > p > a').get('href')
```

```python
div_list[0].select('div.right > p')
```

```python
site_list = []
base_url = 'https://www.alexa.com'
for div in div_list:
    rank = div.find(class_='td').text
    site = div.select_one('div.DescriptionCell > p > a').text
    info = base_url + div.select_one('div.DescriptionCell > p > a').get('href')
    time = div.select('div.right > p')[0].text
    visitor = div.select('div.right > p')[1].text
    traffic = div.select('div.right > p')[2].text
    total = div.select('div.right > p')[3].text
    site_list.append([rank, site, info, time, visitor, traffic, total])
site_list
```

```python
site_df = pd.DataFrame(site_list, columns = ['rank', 'site', 'info', 'time', 'visitor', 'traffic', 'total'])
site_df
```

```python
site_df.to_csv('output/alexa.csv', index = False)
```
