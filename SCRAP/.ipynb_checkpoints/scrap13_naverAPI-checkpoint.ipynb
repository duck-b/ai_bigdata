```python
def search_book(book_title):
    encText = urllib.parse.quote(book_title)

    url = "https://openapi.naver.com/v1/search/book.json?query=" + encText + "&display=50"
    print(url)


    # 페이지 요청 
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()

    # 200 코드이면 response_body에 저장 
    if(rescode==200):
        response_body = response.read()

        book = json.loads(response_body.decode('utf-8'))
        book_list = book['items']
        df_book = pd.DataFrame(columns = ['title', 'link', 'image', 'author', 'price', 'discount', 'publisher', 'pubdate', 'isbn', 'description'])

        idx = 1
        for dict in book_list:
            title = dict['title'].replace('<b>','').replace('</b>','')
            link = dict['link']
            image = dict['image']
            author = dict['author']
            price = dict['price']
            discount = dict['discount']
            publisher = dict['publisher']
            pubdate = dict['pubdate']
            isbn = dict['isbn']
            description = dict['description'].replace('<b>','').replace('</b>','')
            df_book.loc[idx] = [title, link, image, author, price, discount, publisher, pubdate, isbn, description]
            idx += 1

        return df_book
        
```

```python
search_book('해리포터')
```
