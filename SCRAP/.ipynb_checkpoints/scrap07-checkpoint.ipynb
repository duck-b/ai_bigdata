```python
kid_list = []
url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page='
i = 0

while True:
    res = requests.get(url+str(i+1))
    html_str = res.text
    soup = BeautifulSoup(html_str, 'html.parser')
    td_list = soup.select('td.title')
    a_list = soup.select('a.author')

    for j in range(len(a_list)):
        title = td_list[j].find('a').text
        score = td_list[j].find('em').text
        comment = td_list[j].find('br').next_sibling.replace('\n','').replace('\t','')
        writer = a_list[j].text
        date = a_list[j].next_sibling.next_sibling
        kid_list.append([title, score, comment, writer, date])
    
    if soup.find(class_='pg_next'):
        i = i + 1
    else:
        break
    
kid_df = pd.DataFrame(kid_list, columns = ['title', 'score', 'comment', 'writer', 'date'])
kid_df
```

```python
kid_df.to_csv('output/movieKid.csv', index = False)
```

```python
# 함수로 간단히 만들기
def movie_review(url):
    review_list = []
    i = 0

    while True:
        res = requests.get(url+str(i+1))
        html_str = res.text
        soup = BeautifulSoup(html_str, 'html.parser')
        td_list = soup.select('td.title')
        a_list = soup.select('a.author')

        for j in range(len(a_list)):
            title = td_list[j].find('a').text
            score = td_list[j].find('em').text
            comment = td_list[j].find('br').next_sibling.replace('\n','').replace('\t','')
            writer = a_list[j].text
            date = a_list[j].next_sibling.next_sibling
            review_list.append([title, score, comment, writer, date])

        if soup.find(class_='pg_next'):
            i = i + 1
        else:
            break
    
    df = pd.DataFrame(review_list, columns = ['title', 'score', 'comment', 'writer', 'date'])
    return df
```

```python
movie_review('https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page=')
```
