```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://www.yes24.com/Main/default.aspx'
res = requests.get(url)
res
```

```python
html_str = res.text
soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
li_list = soup.select('div.yBestOl > ol > li')
len(li_list)
```

```python
li_list[0]
```

```python
li_list[0].select_one('li > strong.icon').text
```

```python
li_list[0].select_one('li > span > strong').text
```

```python
li_list[0].select_one('li > span > em').text
```

```python
li_list[0].select_one('li > a').get('href')
```

```python
li_list[0].select_one('li > a').get('title').split()[-1]
```

```python
li_list[0].select_one('img').get('src').replace('/S','/XL')
```

```python
book_list = []
base_url = 'http://www.yes24.com'
for li in li_list:
    rank = li.select_one('li > strong.icon').text
    title = li.select_one('li > span > strong').text
    writer = li.select_one('li > span > em').text
    link = base_url + li.select_one('li > a').get('href')
    public = li.select_one('li > a').get('title').split()[-1]
    img = li.select_one('img').get('src').replace('/S','/XL')
    book_list.append([rank, title, writer, link, public, img])
book_list
```

```python
book_df = pd.DataFrame(book_list, columns = ['rank', 'title', 'writer', 'link', 'public', 'img'])
book_df
```

```python
book_df.to_csv('output/yes24.csv', index = False)
```
