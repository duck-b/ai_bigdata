```python
from xml.etree.ElementTree import *
import pandas as pd
import requests
```

```python
url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson'
params ={
    'serviceKey' : 'Yhq6n9vlpaKZsCOw0jec8zIdZ8p+Bpuku6WLjVgDhqXRW6dHnfXoauSEj19jpnjv59CUGspyTxVCgCeXOua7dg==',
    'pageNo' : '1',
    'numOfRows' : '10',
    'startCreateDt' : '20210101',
    'endCreateDt' : '20210630'
}

response = requests.get(url, params=params)
print(response.text)
```

```python
# xml객체명 = fromstring(xml데이터소스: string)
tree = ElementTree(fromstring(response.text))
tree
```

```python
root = tree.getroot()
root, type(root)
```

```python
body = root.find('body')
body
```

```python
items = body.find('items')
items
```

```python
item_list = items.findall('item')
item_list, len(item_list)
```

```python
item_list = root.find('body').find('items').findall('item')
item_list, len(item_list)
```

```python
item_list[0].find('stateDt').tag, item_list[0].find('stateDt').text
```

```python
result_list = []
for item in item_list:
    stateDt = item.find('stateDt').text
    decideCnt = item.find('decideCnt').text
    examCnt = item.find('examCnt').text
    deathCnt = item.find('deathCnt').text
    result_list.append([stateDt, decideCnt, examCnt, deathCnt])
    
result_list
```

```python
df_result = pd.DataFrame(result_list, columns=['stateDt', 'decideCnt', 'examCnt', 'deathCnt'])
df_result.head()
```
