```python
import requests
from bs4 import BeautifulSoup
```

```python
dir(BeautifulSoup)
```

```python
# Sample Scraping
url = 'http://pythonscraping.com/pages/warandpeace.html'
res = requests.get(url)
res
```

```python
html_str = res.text
html_str
```

```python
soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
soup.style
```

```python
soup.h1
```

```python
soup.h1.text
```

```python
print(soup.h1)
print(soup.h1.next_sibling)
print(soup.h1.next_sibling.next_sibling)
print(soup.h1.next_sibling.next_sibling.next_sibling)
```

```python
print(soup.h1.next_sibling.next_sibling.text)
```

```python
print(soup.find('h1'))
print(soup.find('p'))
```

```python
print(soup.find(id='text').text)
```

```python
print(soup.find(class_='red'))
```

```python
# 태그로 전체 검색
span_list = soup.find_all('span')
len(span_list)
```

```python
for item in span_list:
    print(item.text)
    print('='*50)
```

```python
# 클래스로 전체 검색
red_list = soup.find_all(class_='red')
len(red_list)
```

```python
for item in red_list:
    print(item.text)
    print('='*50)
```

```python
green = soup.find(class_='green')
green
```

```python
# 속성 값 출력
green.get('class')
```
