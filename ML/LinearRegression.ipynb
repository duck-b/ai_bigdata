```python
## Linear Regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
```

```python
DF = load_diabetes()
print(DF.DESCR)
```

```python
dir(DF)
```

```python
DF.data_filename
```

```python
DF.feature_names
```

```python
X = pd.DataFrame(DF.data, columns=DF.feature_names)
X
```

```python
y = pd.DataFrame(DF.target, columns=['target'])
print(y.shape)
display(y.head())
display(y.tail())
```

```python
train_X = np.array(X[:-20]['bmi']).reshape(-1,1)
display(train_X)
len(train_X)
```

```python
test_X = np.array(X[-20:]['bmi']).reshape(-1,1)
display(test_X)
len(test_X)
```

```python
train_y = y[:-20]
test_y = y[-20:]
```

```python
regr = LinearRegression()
regr.fit(train_X, train_y)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
plt.scatter(test_X, test_y, color='black')
plt.plot(test_X, regr.predict(test_X), color='b')
plt.show()
```
