```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc

rc('font', family='AppleGothic')
matplotlib.rcParams['axes.unicode_minus'] = False

plt.style.use('ggplot')
```

```python
df = pd.read_excel('./data/남북한발전전력량.xlsx', convert_float=True)
df = df.loc[5:9]
df
```

```python
df.drop('전력량 (억㎾h)', axis=1, inplace=True)
df.set_index('발전 전력별', inplace=True)
df
```

```python
df = df.T
df
```

```python
df['합계 - 1년'] = df['합계'].shift(1)
df
```

```python
df['증감율'] = ((df['합계']/df['합계 - 1년']) - 1) * 100
df
```

```python
ax1 = df[['수력', '화력']].plot(kind='bar', figsize=(20,10), width=0.7, stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감율, ls='--', marker='o', markersize=20, color='green', label='전년대비 증감율(%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량 (억KWh)')
ax2.set_ylabel('전년 대비 증감율(%)')

plt.title('북한 전력 발전량(1990 ~ 2016)', size=30)
ax1.legend(loc='upper left')

plt.show()
```

```python
plt.style.use('default')

df = pd.read_csv('./data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

cylinders_size = (df.cylinders / df.cylinders.max()*5) ** 3

df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=cylinders_size, figsize=(10,5), alpha=0.3)
plt.title('Scatter Plot - mpg vs. weight')
plt.show()
```

```python
df.plot(kind='scatter', x='weight', y='mpg', marker='+', figsize=(10,5),
       cmap='viridis', c=cylinders_size, s=50, alpha=0.3)
plt.title('Scatter Plot - mpg vs. weight')
plt.show()
```

```python
plt.style.use('seaborn-poster')

df = pd.read_csv('./data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.boxplot(x=[
                df[df['origin'] == 1]['mpg'], 
                df[df['origin'] == 2]['mpg'], 
                df[df['origin'] == 3]['mpg']
            ], labels=['USA', 'EU', 'JAPAN']
)

ax2.boxplot(x=[
                df[df['origin'] == 1]['mpg'], 
                df[df['origin'] == 2]['mpg'], 
                df[df['origin'] == 3]['mpg']
            ], labels=['USA', 'EU', 'JAPAN'], vert=False
)

plt.show()
```

```python
X = np.random.normal(0,1,100)
Y = np.random.normal(0,1,100)
C = np.random.randint(0,5,100)

cmap_list = [plt.cm.rainbow, plt.cm.Blues, plt.cm.autumn, plt.cm.RdYlGn]

f, axes = plt.subplots(1,4, sharex=True, sharey=True)
f.set_size_inches((16,4))
for i in range(0,4):
    axes[i].scatter(X, Y, c=C, cmap=cmap_list[i])
    axes[i].set_title("cmap: {}".format(cmap_list[i].name))
    
plt.show()
```

```python
arr = np.random.standard_normal((8,100))

plt.subplot(2,2,1)
plt.scatter(arr[0],arr[1], c=arr[1])
plt.spring()
plt.title('spring')

plt.subplot(2,2,2)
plt.scatter(arr[0],arr[1], c=arr[1])
plt.summer()
plt.title('summer')

plt.subplot(2,2,3)
plt.scatter(arr[0],arr[1], c=arr[1])
plt.autumn()
plt.title('autumn')

plt.subplot(2,2,4)
plt.scatter(arr[0],arr[1], c=arr[1])
plt.winter()
plt.title('winter')
```
