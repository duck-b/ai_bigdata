```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [69.0, 80.0, 257.0, 1686.0, 6505.0, 11865.3, 22105.3]

plt.bar(range(len(years)), gdp,
       color=['r','g','b','yellow','pink','magenta','lime'])
plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xticks(range(len(years)), years)
plt.show()
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x_range = np.arange(len(years))
plt.bar(x_range - 0.3, ko, width=0.3, label='Korea')
plt.bar(x_range + 0.0, jp, width=0.3, label='Japan')
plt.bar(x_range + 0.3, ch, width=0.3, label='China')

plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xlabel('years')
plt.legend()
plt.xticks(x_range, years)
plt.show()
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x_range = np.arange(len(years))
plt.bar(x_range, ko, label='Korea')
plt.bar(x_range, jp, label='Japan', bottom=pd.Series(ko))
plt.bar(x_range, ch, label='China', bottom=pd.Series(ko)+pd.Series(jp))

plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xlabel('years')
plt.legend()
plt.xticks(x_range, years)
plt.show()
```

```python
import warnings
with warnings.catch_warnings(record=True):
    warnings.simplefilter('always')
    df = pd.read_excel('./data/시도별 전출입 인구수.xlsx', engine='openpyxl', header=0)
df.head()
```

```python
# 결측치를 앞에값으로 채움
df = df.fillna(method='ffill')
df
```

```python
lst = np.array([1,2,3,5])
mask = [False, True, True, False]
lst[mask]
```

```python
df['전출지별']
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')

df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별': '전입지'}, axis=1, inplace=True)
df_seoul.set_index(['전입지'], inplace=True)

df_seoul
```

```python
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.title('서울->경기 인구이동')

plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.legend(labels=['서울->경기'], loc='upper left')
plt.show()
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(14,5))
plt.xticks(size=10, rotation=45)
plt.plot(sr_one.index, sr_one.values, marker='o', ms=5)
plt.title('서울 -> 경기 인구 이동', size=30)
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)
plt.legend(labels=['서울->경기'], loc='best', fontsize=14)
plt.show()
```

```python
plt.style.use('ggplot')

# 그래프 크기
plt.figure(figsize=(14,5))

# x축 눈금
plt.xticks(size=10, rotation=45)

# 그래프
plt.plot(sr_one.index, sr_one.values, marker='o', ms=5)

# 그래프 제목
plt.title('서울 -> 경기 인구 이동', size=30)

# x, y축 이름
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

# 범주 표시
plt.legend(labels=['서울->경기'], loc='best', fontsize=14)

# y 축 범위 지정
plt.ylim(50000, 800000)

# 주석 - 화살표
plt.annotate('',
             xy=(20,600000), # 화살표의 머리 부분(x인덱스, y좌표)
             xytext=(2,290000), # 화살표의 꼬리 부분(x인덱스, y좌표)
             xycoords='data', # 좌표체계
             arrowprops=dict(arrowstyle='->', color='red', lw=5) # 화살표 서식
)

plt.annotate('',
             xy=(47,450000), # 화살표의 머리 부분(x인덱스, y좌표)
             xytext=(30,580000), # 화살표의 꼬리 부분(x인덱스, y좌표)
             xycoords='data', # 좌표체계
             arrowprops=dict(arrowstyle='->', color='blue', lw=5) # 화살표 서식
)

plt.annotate('인구이동 증가(1970-1995)',
             xy=(10,380000),
             rotation=22,
             va='baseline',
             ha='center',
             fontsize=15
)

plt.annotate('인구이동 감소(1995-2017)',
             xy=(39,510000),
             rotation=-9.5,
             va='baseline',
             ha='center',
             fontsize=15
)

plt.show()
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(sr_one, marker='o', markerfacecolor='orange', markersize=10, color='olive', label='경기')
ax.legend(loc='best')
ax.set_ylim(5000, 800000)
ax.set_title('서울 => 경기 인구 이동', size=20)
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)
ax.set_xticklabels(sr_one.index, rotation=75)

plt.show()
```

```python
col_years = list(map(str, df_seoul.columns))

df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도'], col_years]
df_3
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, df_3.loc['충청남도', :],
        marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='충남')

ax.plot(col_years, df_3.loc['경상북도', :],
        marker='o', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='경북')

ax.plot(col_years, df_3.loc['강원도', :],
        marker='o', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='강원')

ax.legend(loc='best')

ax.set_title('서울 -> 충남, 경북, 강원 인구 이동', size=25)
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)
ax.set_xticklabels(col_years, rotation=45)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

plt.show()
```

```python
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]

plt.style.use('ggplot')
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(col_years, df_4.loc['충청남도', :],
        marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='충남')

ax2.plot(col_years, df_4.loc['경상북도', :],
        marker='o', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='경북')

ax3.plot(col_years, df_4.loc['강원도', :],
        marker='o', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='강원')

ax4.plot(col_years, df_4.loc['전라남도', :],
        marker='o', markerfacecolor='yellow', markersize=10,
        color='orange', linewidth=2, label='전남')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

plt.show()
```
