```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [69.0, 80.0, 257.0, 1686.0, 6505.0, 11865.3, 22105.3]

plt.bar(range(len(years)), gdp,
       color=['r','g','b','yellow','pink','magenta','lime'])
plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xticks(range(len(years)), years)
plt.show()
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x_range = np.arange(len(years))
plt.bar(x_range - 0.3, ko, width=0.3, label='Korea')
plt.bar(x_range + 0.0, jp, width=0.3, label='Japan')
plt.bar(x_range + 0.3, ch, width=0.3, label='China')

plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xlabel('years')
plt.legend()
plt.xticks(x_range, years)
plt.show()
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x_range = np.arange(len(years))
plt.bar(x_range, ko, label='Korea')
plt.bar(x_range, jp, label='Japan', bottom=pd.Series(ko))
plt.bar(x_range, ch, label='China', bottom=pd.Series(ko)+pd.Series(jp))

plt.title('GDP per capita')
plt.ylabel('dollars')
plt.xlabel('years')
plt.legend()
plt.xticks(x_range, years)
plt.show()
```

```python
df['']
mask = (df[''] == '') & (df[''] != '')

df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis=1)
df_seoul.rename({'': ''}, axis=1, inplace=True)
df_seoul.set_index([''], inplace=True)

df_seoul
```

```python
sr_one = df_seoul.loc['']
sr_one
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.title('-> ')

plt.xlabel('')
plt.ylabel(' ')

plt.legend(labels=['->'], loc='upper left')
plt.show()
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(14,5))
plt.xticks(size=10, rotation=45)
plt.plot(sr_one.index, sr_one.values, marker='o', ms=5)
plt.title(' ->   ', size=30)
plt.xlabel('', size=20)
plt.ylabel(' ', size=20)
plt.legend(labels=['->'], loc='best', fontsize=14)
plt.show()
```

```python
plt.style.use('ggplot')

#  
plt.figure(figsize=(14,5))

# x 
plt.xticks(size=10, rotation=45)

# 
plt.plot(sr_one.index, sr_one.values, marker='o', ms=5)

#  
plt.title(' ->   ', size=30)

# x, y 
plt.xlabel('', size=20)
plt.ylabel(' ', size=20)

#  
plt.legend(labels=['->'], loc='best', fontsize=14)

# y   
plt.ylim(50000, 800000)

#  - 
plt.annotate('',
             xy=(20,600000), #   (x, y)
             xytext=(2,290000), #   (x, y)
             xycoords='data', # 
             arrowprops=dict(arrowstyle='->', color='red', lw=5) #  
)

plt.annotate('',
             xy=(47,450000), #   (x, y)
             xytext=(30,580000), #   (x, y)
             xycoords='data', # 
             arrowprops=dict(arrowstyle='->', color='blue', lw=5) #  
)

plt.annotate(' (1970-1995)',
             xy=(10,380000),
             rotation=22,
             va='baseline',
             ha='center',
             fontsize=15
)

plt.annotate(' (1995-2017)',
             xy=(39,510000),
             rotation=-9.5,
             va='baseline',
             ha='center',
             fontsize=15
)

plt.show()
```

```python
col_years = list(map(str, df_seoul.columns))

df_3 = df_seoul.loc[['', '', ''], col_years]
df_3
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, df_3.loc['', :],
        marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='')

ax.plot(col_years, df_3.loc['', :],
        marker='o', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='')

ax.plot(col_years, df_3.loc['', :],
        marker='o', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='')

ax.legend(loc='best')

ax.set_title(' -> , ,   ', size=25)
ax.set_xlabel('', size=12)
ax.set_ylabel(' ', size=12)
ax.set_xticklabels(col_years, rotation=45)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

plt.show()
```

```python
df_4 = df_seoul.loc[['', '', '', ''], col_years]

plt.style.use('ggplot')
fig = plt.figure(figsize=(40,20))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(col_years, df_4.loc['', :],
        marker='o', markerfacecolor='green', markersize=10,
        color='olive', linewidth=2, label='')

ax2.plot(col_years, df_4.loc['', :],
        marker='o', markerfacecolor='blue', markersize=10,
        color='skyblue', linewidth=2, label='')

ax3.plot(col_years, df_4.loc['', :],
        marker='o', markerfacecolor='red', markersize=10,
        color='magenta', linewidth=2, label='')

ax4.plot(col_years, df_4.loc['', :],
        marker='o', markerfacecolor='yellow', markersize=10,
        color='orange', linewidth=2, label='')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_title(' ->   ', size=15)
ax2.set_title(' ->   ', size=15)
ax3.set_title(' ->   ', size=15)
ax4.set_title(' ->   ', size=15)

ax1.set_xticklabels(col_years, rotation=45)
ax2.set_xticklabels(col_years, rotation=45)
ax3.set_xticklabels(col_years, rotation=45)
ax4.set_xticklabels(col_years, rotation=45)

plt.show()
```
