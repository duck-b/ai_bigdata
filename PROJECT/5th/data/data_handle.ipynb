```python
import pandas as pd
```

```python
comment = pd.read_csv('./m_coment.csv', encoding='CP949')
```

```python
comment['user_id'] = comment['user_id'].str.replace(',','')
```

```python
comment['text'] = comment['text'].str.replace(',','')
```

```python
comment = comment[['movie_id', 'comment_userId', 'comment_userName', 'comment_point', 'comment_contents']]
```
