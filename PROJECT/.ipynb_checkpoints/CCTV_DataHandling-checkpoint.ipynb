```python
import pandas as pd
import numpy as np
# pd.set_option('display.max_rows', 100)
```

```python
cctv = pd.read_excel('./CCTV_부산.xlsx')
cctv
```

```python
cctv1 = cctv.loc[:,['관리기관명', '설치목적구분', '설치연월', 'WGS84위도', 'WGS84경도', '데이터기준일자']]
cctv1
```

```python
pd.DataFrame(cctv1['설치연월'].value_counts())
```

```python
cctv1['설치연월'] = cctv1['설치연월'].fillna('2016-01')
```

```python
cctv1
```

```python
cctv2_ym = cctv1['설치연월'].str.split('-')
cctv2_year = []
for i in range(len(cctv2_ym)):
    cctv2_year.append(cctv2_ym[i][0])
cctv2 = cctv1.copy()
cctv2['설치연월'] = cctv2_year
cctv2.rename(columns={'설치연월': '설치년도'}, inplace=True)
cctv2
```

```python
pd.DataFrame(cctv2['관리기관명'].value_counts())
```

```python
cctv3 = cctv2.copy()
cctv3['관리기관명'] = cctv3['관리기관명'].str.replace('부산', '')
cctv3['관리기관명'] = cctv3['관리기관명'].str.replace('광역시', '')
pd.DataFrame(cctv3['관리기관명'].value_counts())
```

```python
cctv3 = cctv3[cctv3['관리기관명'] != '']
cctv3 = cctv3[cctv3['관리기관명'] != ' 남부교육지원청']
```

```python
cctv_data = pd.DataFrame(cctv3[['관리기관명', '설치년도']].value_counts())
cctv_data.sort_values(by=['관리기관명', '설치년도'])
```

```python
cctv_data.columns = ['CCTV수']
cctv_data.sort_values(by=['관리기관명', '설치년도'], inplace=True)
```

```python
cctv_data
```

```python
cctv_data.to_csv('./CCTV_설치년도별.csv', index=False,encoding='utf-8')
```
