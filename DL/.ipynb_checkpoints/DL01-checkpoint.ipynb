```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.array([1,2,3,4,5])
Y = np.array([1,2,3,4,5])

model = Sequential()
model.add(Dense(1, input_dim=1, activation='linear'))
sgd=optimizers.SGD(lr=0.01)
model.compile(optimizer=sgd, loss='mse', metrics=['accuracy'])
model.fit(X, Y, batch_size=1, epochs=10, shuffle=False)

plt.plot(X, Y, 'bo')
plt.plot(X, model.predict(X), 'r-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w):
    f = 0.5 * (w*2-2)**2
    return f

def Df(w):
    df = 2*(w * 2 - 2)
    return df
    
def xp1(w, alpha):
    xp1 = w - alpha * Df(w)
    return xp1

def plot_steps(guess, alpha, nsteps):
    fig, ax = plt.subplots()
    w = np.linspace(-1,3,100)
    ax.plot(w , f(w))
    
    w = guess
    ax.plot(w, f(w), 'o', label='start w = %.2f'%w)
    
    for i in range(nsteps):
        wold = w
        w = xp1(w, alpha)
        ax.plot(w, f(w), 'o')
        ax.plot([wold, w], [f(wold), f(w)], '-')
    
    
    plt.legend()
    plt.show()
    
    
plot_steps(3, 0.01, 80)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    f = 0.5*x**2
    return f

def Df(x):
    df = x*2
    return df

def xp1(x, alpha):
    xp1 = x - alpha * Df(x)
    return xp1

def plot_steps(guess, alpha, nsteps):
    fig, ax = plt.subplots()
    x = np.linspace(-3, 3, 100)
    ax.plot(x, f(x))
    
    x = guess
    ax.plot(x, f(x), 'o', label='start x=%.2f' % x)
    for i in range(nsteps):
        xold = x
        x = xp1(x, alpha)
        ax.plot(x, f(x), 'o')
        ax.plot([xold, x], [f(xold), f(x)], '-')
    plt.legend()
    plt.show()
    
plot_steps(3, 0.03, 80)    
```

```python
import matplotlib.pyplot as plt
import numpy as np

def J(a0, a1, x, y, m):
    ret = 0
    for i in range(m):
        ret += 0.5*((a0+a1*x[i])-y[i])**2
    return ret/m

x = np.linspace(-1,1,5)
y = x
a0 = np.linspace(-3,3,100)
a1 = np.linspace(-3,3,100)
aa0, aa1 = np.meshgrid(a0, a1)
plt.contour(aa0, aa1, J(aa0, aa1, x, y, m=len(x)), colors='C0', levels=[i for i in np.arange(0,80,0.3)])
plt.show()
```
